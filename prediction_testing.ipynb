### Pick what model would fit the best my data

**While trying to find the best solution I want to see what would be the best performance dataset setup.**

#### Create a function that will compare different models and help me pick the best model and data setup

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def evaluate_models(X_train, X_test, y_train, y_test):
    # The models I want to check
    models = {
        'Linear Regression': LinearRegression(),
        'Decision Tree': DecisionTreeRegressor(),
        'Random Forest': RandomForestRegressor(),
        'Gradient Boosting Regressor' : GradientBoostingRegressor(),
        'Ada Boost Regressor' : AdaBoostRegressor()
    }
    
    results = {}
    
    for name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        # Create predictions
        y_pred = model.predict(X_test)
        # Calculate metrics for the model
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        
        results[name] = {'RMSE': rmse, 'R2': r2}
        
    return results

# 2 subplots: RMSE and R2
def view_rmse_r2(results, title):
    models = list(results.keys())
    rmse_values = [results[model]['RMSE'] for model in models]
    r2_values = [results[model]['R2'] for model in models]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
    
    # RMSE
    ax1.bar(models, rmse_values, color='tab:orange', alpha=0.7)
    ax1.set_title('RMSE for each model' + title)
    ax1.set_ylabel('RMSE')
    ax1.set_xlabel('Model')
    ax1.tick_params(axis='x', rotation=45)
    ax1.set_ylim(0, 100)
    
    # Grafic pentru R2
    ax2.bar(models, r2_values, color='tab:blue', alpha=0.7)
    ax2.set_title('R2 for each model' + title)
    ax2.set_ylabel('R2 Score')
    ax2.set_xlabel('Model')
    ax2.tick_params(axis='x', rotation=45)
    ax2.set_ylim(-1, 1)
    
    plt.tight_layout()
    plt.show()
```

#### **Performance for the imported 1:1 data**

```python
data_path = 'data/data.csv'
data_df = pd.read_csv(data_path)

X_train, X_test, y_train, y_test = train_test_split(data_df.drop(['Cultivar', 'MHG'], axis=1), data_df['MHG'], test_size=0.2, random_state=42)
results = evaluate_models(X_train, X_test, y_train, y_test)
for x in results.items():
    print (x)
view_rmse_r2(results, ' for 1:1 data')
```

#### **Does the aggregation of the weather conditions add more accuracy?**

```python
data_path = 'data/data.csv'
data_df = pd.read_csv(data_path)

data = {
    'Month': ['November', 'December', 'January', 'February', 'March', 'April'],
    'Precipitation_2022_23_mm': [365, 600, 560, 365, 520, 750],
    'Precipitation_1991_2020_mm': [350, 295, 310, 405, 430, 405],
    'Temperature_2022_23_C': [25.55, 25.75, 25.1, 25.2, 26.25, 24.59],
    'Temperature_1991_2020_C': [25.49, 25.5, 25.5, 25.12, 24.55, 25.12]
}

weather_df = pd.DataFrame(data)

climate_mapping = weather_df.set_index('Month').T.to_dict('list')
for i in range(1, 6):  # 5 months for every cultivation season
    data_df[f'temp_month{i}'] = 0.0
    data_df[f'pp_month{i}'] = 0.0

for index, row in data_df.iterrows():
    if row['Season'] == 1:
        # season 1: November - March
        months = ['November', 'December', 'January', 'February', 'March']
    else:
        # season 2: December - April
        months = ['December', 'January', 'February', 'March', 'April']
    
    for i, month in enumerate(months, start=1):
        data_df.at[index, f'pp_month{i}'] = climate_mapping[month][0]  # Index 0 for precipitation
        data_df.at[index, f'temp_month{i}'] = climate_mapping[month][2]  # Index 2 for temperature

X_train, X_test, y_train, y_test = train_test_split(data_df.drop(['Cultivar', 'MHG'], axis=1), data_df['MHG'], test_size=0.2, random_state=42)
results = evaluate_models(X_train, X_test, y_train, y_test)
for x in results.items():
    print (x)
view_rmse_r2(results, ' with weather')
```

**For some models the aggregation of the weather conditions adds more accuracy, but not that much.**

### Training and hyperparametrization

```python
data_path = 'data/data.csv'
data_df = pd.read_csv(data_path)

X_train, X_test, y_train, y_test = train_test_split(data_df.drop(['Cultivar', 'MHG'], axis=1), data_df['MHG'], test_size=0.2, random_state=42)

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# define Randomized Search space 
random_search_space = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_features': [1.0, 'sqrt'], 
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [ 2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Initialize and fit Randomized Search
random_search = RandomizedSearchCV(estimator=RandomForestRegressor(), param_distributions=random_search_space, n_iter=200, cv=5, verbose=2, random_state=42, n_jobs=-1)
random_search.fit(X_train, y_train)
```

```python
best_random_params = random_search.best_params_ 
```

```python
best_random_params
```

```python
y_pred = random_search.best_estimator_.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"Score R^2 for the best model: {r2}")
```

```python
grid_search_space = {
    'n_estimators': [280, 300, 320], 
    'max_depth': [30, 35, 40, 45, 50],  
    'min_samples_split': [2],  
    'min_samples_leaf': [1], 
    'max_features': [0.9, 0.95, 1.0]  
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(), param_grid=grid_search_space, cv=5, verbose=2, n_jobs=-1)

grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_ 
```

```python
best_params
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"Score R^2 for the best model: {r2}")
```
